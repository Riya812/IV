Color image

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from PIL import Image
```

```python
img_clr = cv2.imread('IMG_20190414_104624.jpg')

plt.imshow(img_clr)
```

```python
print('Shape of the image:', img_clr.shape)
print('Size of the image:', img_clr.size)
```

```python
new_size = (400, 800)
resize_img_clr = cv2.resize(img_clr, new_size)

plt.imshow(resize_img_clr)
```

```python
print('Shape of the image:', resize_img_clr.shape)
print('Size of the image:', resize_img_clr.size)
```

```python
img = Image.fromarray(resize_img_clr.astype('uint8'))  # Ensure data type is uint8

# Save the image
img.save('IMG_20190414_104624_1.jpg')


copied_image = Image.open('IMG_20190414_104624_1.jpg')
plt.imshow(copied_image)
plt.axis('off')  # Hide axes
plt.show()
```
