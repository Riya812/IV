```python
import cv2  # OpenCV library for image handlin
import matplotlib.pyplot as plt  # For displaying images inline
```

```python
# Read the image in BGR format
image = cv2.imread('20230128_180019.jpg')

# Convert to RGB for proper color display in matplotlib
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

image
```

```python
# Display the image using matplotlib
plt.imshow(image)
plt.axis('off')  # Hide axes for better visualization
plt.show()

```

```python
# Save the image as a new file
new_image = cv2.imwrite('IMG_20190414_104624.jpg', image)

print("Image type:", type(image))

print("Image shape:", image.shape)

print("Image size:", image.size)
```
