```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
image = cv2.imread('photo-1606115915090-be18fea23ec7.tiff')

plt.imshow(image)
```

```python
image1 = cv2.imread('IMG_20190414_104624.jpg')
plt.imshow(image1)
```

```python
#lossless(PNG)

cv2.imwrite('image_lossless.png', image)
cv2.imwrite('image_lossless1.png', image1)


#lossy(JPEG)

cv2.imwrite('image_lossy.jpg', image, [cv2.IMWRITE_JPEG_QUALITY, 90])
cv2.imwrite('image_lossy1.jpg', image1, [cv2.IMWRITE_JPEG_QUALITY, 90])
 
```

```python
#Image

image_lossless = cv2.imread('image_lossless.png')

image_lossy = cv2.imread('image_lossy.jpg')

#plot
plt.subplot(1,3,1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('original')
plt.show()

plt.subplot(1,3,2)
plt.imshow(cv2.cvtColor(image_lossless, cv2.COLOR_BGR2RGB))
plt.title('Lossless I')
plt.show()

plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(image_lossy, cv2.COLOR_BGR2RGB))
plt.title('Lossy I')
plt.show()
```

```python
#Image1

image_lossless1 = cv2.imread('image_lossless1.png')

image_lossy1 = cv2.imread('image_lossy1.jpg')

#plot
plt.subplot(1,3,1)
plt.imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2RGB))
plt.title('original 2')
plt.show()

plt.subplot(1,3,2)
plt.imshow(cv2.cvtColor(image_lossless1, cv2.COLOR_BGR2RGB))
plt.title('Lossless II')
plt.show()

plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(image_lossy1, cv2.COLOR_BGR2RGB))
plt.title('Lossy II')
plt.show()
```
