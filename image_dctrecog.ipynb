```python
# Low thresholds for more edges (more noise)
edges_low = cv2.Canny(gray, 50, 100)

# High thresholds for fewer, clearer edges
edges_high = cv2.Canny(gray, 150, 200)

# Display them to compare
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(edges_low, cmap='gray')
plt.title('Low Thresholds (50, 100)')

plt.subplot(1, 2, 2)
plt.imshow(edges_high, cmap='gray')
plt.title('High Thresholds (150, 200)')

plt.show()
```

```python
dct1 = cv2.dct(np.float32(gray))

plt.figure(figsize = (5,5))
plt.imshow(dct1, cmap = 'gray')
plt.title('Discrete Cosine Transform')
plt.show()
```

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')

faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors= 5, minSize = (30,30))

```

```python
for (x,y,w,h) in faces:
    cv2.rectangle(image, (x,y),(x+w,y+h),(0,255,0),20)

plt.figure(figsize=(10,5))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Face Detection')
plt.show()


for (x,y,w,h) in faces:
    cv2.rectangle(gray, (x,y),(x+w,y+h),(0,255,0),20)

plt.figure(figsize=(10,5))
plt.imshow(cv2.cvtColor(gray, cv2.COLOR_BGR2RGB))
plt.title('Face Detection')
plt.show()
```
