```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = np.zeros((512,512,3), np.uint8)
```

```python
cv2.line(image, (0,0),(500,500),(255,255,255),5)

plt.imshow(image)
```

```python
image = np.zeros((500,500,3), np.uint8)
cv2.arrowedLine(image, (500,0),(0,500),(255,255,255),5)
plt.imshow(image)
```

```python
image = np.zeros((500,500,3),dtype = "uint8")

cv2.rectangle(image,(100,100),(400,400),(255,255,255),5)
plt.imshow(image)
```

```python
image = np.zeros((500,500,3), dtype = "uint8")

cv2.circle(image, (250,250),100,(255,255,255),5)
plt.imshow(image)
```

```python
image = np.zeros((512,512,3), dtype = "uint8")

cv2.putText(image, 'RIYA JADHAV', (50,450), cv2.FONT_HERSHEY_SIMPLEX, 2, (255,255,255),5)

plt.imshow(image)
```
